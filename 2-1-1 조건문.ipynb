## 1. import dataframe 

### 1. import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. import mtcars data

```python
data = pd.read_csv('https://gist.githubusercontent.com/ZeccaLehn/4e06d2575eb9589dbe8c365d61cb056c/raw/64f1660f38ef523b2a1a13be77b002b98665cdfe/mtcars.csv')
data.rename(columns={'Unnamed: 0':'brand'}, inplace=True)
```

```python
sample_data = data
```

## 3. if/else to create column
 mpg >20 big ,else small

```python
### 1. 한 조건만 바꿀 때 쓰는 법(loc) ###
sample_data.loc[(sample_data.mpg>20),'size'] = 'big'
```

```python
### 2. if-else
sample_data['size'] = ['big' if x > 20 else 'small' for x in sample_data['mpg']]
```

```python
### 3. where
sample_data['size'] = 0
sample_data['size'] = np.where(sample_data['mpg']>20,'big', 'small')
```

```python
### 3. 단순 if-else 구문 오류
#    if data["mpg"]>20 :
#        data["size"] = 'big'
#    else :
#        data["size"] = "small"

# 위의 경우는 boolean error가 발생하므로 반드시 위의 경우로 해결
```

## 4. if/else to change value
### brand name is Merc hp = hp * 2

```python
### 1. 한조건만 바꿀때 (loc)
sample_data.loc[sample_data['brand'].str[0:4] == 'Merc', 'hp'] = sample_data['hp']*2
sample_data.loc[sample_data['brand'].str[0:4] != 'Merc', 'hp'] = sample_data['hp']
```

```python
### 2. if-else using lambda
sample_data['hp2'] = sample_data['brand'].apply(lambda z : 2 if z[0:4] == 'Merc' else 1) * sample_data['hp']
```

## 5. if/elif/else to create column
### vs:0 am:0 bin:a
### vs:0 am:1 bin:b
### vs:1 am:0 bin:c
### vs:1 am:1 bin:d

```python
### 1. .loc and 대신 & 괄호도 무조건 넣어주기
sample_data.loc[(sample_data['vs'] == 0) & (sample_data['am'] == 0),'bin'] = 'a'
sample_data.loc[(sample_data['vs'] == 0) & (sample_data['am'] == 1),'bin'] = 'b'
sample_data.loc[(sample_data['vs'] == 1) & (sample_data['am'] == 0),'bin'] = 'c'
sample_data.loc[(sample_data['vs'] == 1) & (sample_data['am'] == 1),'bin'] = 'd'
```
